```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('StudentPerformanceFactors.csv')
data.head()
```

```python
features = data.columns
```

```python
extra_curricular = data.groupby('Extracurricular_Activities').Exam_Score.mean()
extra_curricular
```

```python
parents_involvement = data.groupby('Parental_Involvement').Exam_Score.mean()
parents_involvement
```

```python
study_hours = data.groupby('Hours_Studied').Exam_Score.mean()
study_hours
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)  # Create the first subplot
plt.scatter(data['Hours_Studied'], data['Exam_Score'], alpha=0.5, color='green')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.title('Exam Score vs Hours Studied')

plt.subplot(1, 2, 2)  # Create the second subplot
plt.scatter(data['Sleep_Hours'], data['Exam_Score'], alpha=0.5, color='black')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Sleep Hours')
plt.grid(True)
plt.show()
```

```python
scores_correlation = data['Previous_Scores'].corr(data['Exam_Score'])
print(f"Correlation between Previous Scores and Exam Scores: {scores_correlation}")
```

```python
plt.scatter(data['Previous_Scores'], data['Exam_Score'], alpha=0.4)
plt.xlabel('Previous Score')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Previous Score')
plt.grid(True)
plt.show()
```
